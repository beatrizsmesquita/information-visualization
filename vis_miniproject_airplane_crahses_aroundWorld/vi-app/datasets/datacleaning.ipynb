```python
# Necessary Libraries

import pandas as pd
import numpy as np
```

```python
#Loading data

data = pd.read_csv('Airplane_Crashes_and_Fatalities_Since_1908.csv')
data
```

### Drop unnecessary columns

```python
# Selecting info - dropping unnecessary columns
data = data.drop(['Time'], axis=1)
data = data.drop(['Flight #'], axis=1)
data = data.drop(['Route'], axis=1)
data = data.drop(['Type'], axis=1)
data = data.drop(['Registration'], axis=1)
data = data.drop(['cn/In'], axis=1)
data = data.drop(['Summary'], axis=1)
data
#Dropped 7 columns stayed with 6
```

### Duplicated rows

```python
# Check if there are duplicated rows

duplicated = data[data.duplicated(keep = False)]
duplicated
```

```python
# There are no duplicated rows
```

### Replacing "Date" columns with "Year" column

```python
# Creating a new column called year just with the years

data['Year'] = pd.DatetimeIndex(data['Date']).year

# Dropping date column

data = data.drop(['Date'], axis=1)
data
```

### Total fatalities per year

```python
groups = data.groupby(data['Year'])
total_fatalitiesdata = groups.sum()
total_fatalitiesdata = total_fatalitiesdata[['Fatalities']].copy()
total_fatalitiesdata
```

### Total crashes per year

```python
total_crashes = data['Year'].value_counts()
totcrashes_data=total_crashes.to_frame()
totcrashes_data.rename(columns = {'Year':'Total Crashes'}, inplace = True)
totcrashes_data.index.name = 'Year'
totcrashes_data
```

```python
timelinedata = pd.concat([total_fatalitiesdata, totcrashes_data],axis=1)
timelinedata
```

```python
timelinedata.to_csv('timeline-copy.csv') 
```

# Operator by Type (%)

```python
Temp = data.copy()
Temp['OperatorType'] = Temp.Operator.str.contains('Military')
Temp = ((Temp.groupby('OperatorType')[['OperatorType']].count())/5268)*100
Temp.index = ['NonMilitary', 'Military']
Temp = Temp.transpose()
Temp
 #Here the sum of military and non military is 5250 beacuse we dont count empty lines
```

```python
Temp2 = data.copy()
Temp2['OperatorType'] = Temp2.Operator.str.contains('Private')
Temp2 = ((Temp2.groupby('OperatorType')[['OperatorType']].count())/5268)*100
Temp2.index = ['NonPrivate', 'Private']
Temp2 = Temp2.transpose()
Temp2
#Here the sum of private and non private is 5250 beacuse we dont count empty lines
```

```python
Temp3 = data.copy()
Temp3['Military'] = Temp3.Operator.str.contains('Military')
Temp3['Private'] = Temp3.Operator.str.contains('Private')

Temp3['OperatorType'] = (Temp3.Military == False) & (Temp3.Private == False)
Temp3 = ((Temp3.groupby('OperatorType')[['OperatorType']].count())/5268)*100
Temp3.index = ['NonOthers', 'Others']
Temp3 = Temp3.transpose()
Temp3
#This counts the empty lines as others
```

```python
df_operators = pd.concat([Temp, Temp2, Temp3],axis=1)
df_operators = df_operators.drop(['NonMilitary', 'NonPrivate', 'NonOthers'], axis=1)
df_operators
```

```python
df_operators.to_csv("operatortype.csv")
```
